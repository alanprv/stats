## Chapter 2: Random Variables

### Exercise 13
Let X ~ N (0,1) and let Y = e^X

(a) Find pdf for Y and plot it.

```python
n = 10000
y = np.zeros(n)
f = np.zeros(n)

for i in range(1,n):
    
    y[i] = i/1000
    f[i] = (1 / (np.sqrt(2 * np.pi) * y[i])) * np.exp((-np.log(y[i]) **2) / 2)
    
plt.plot(y, f)
```

## Exercise 15
Write a program that takes Uniform (0,1) random variables and generates random variables from an Exponential (beta) distribution

```python
beta = 2
n_samples = 10000

def generate(beta, loc=0, scale=1):
    u = np.random.uniform(loc, scale, n_samples)
    return -beta * np.log(1 - u)

exponential = generate(beta)

plt.hist(exponential, bins=300, edgecolor='black');
```
