### Exercise 9
Let $X_1, X_2 ... X_n$ be N(0,1) random variables and let $\overline{X_n}$ = $n^{-1} \sum_{i=1}^n(X_i)$. Plot $\overline{X_n}$ versus n for n=1, 1000. Repeat for $X_1, X_2 ... X_n$  ~ Cauchy. Explain why there's such a difference



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, uniform, cauchy, bernoulli
```

First we do normal distribution

```python
max_n = 10000
x_mean = np.zeros(max_n)
n = np.linspace(1, max_n, max_n)

for i in range(max_n):
    x = norm.rvs(size = i)
    x_mean[i] = np.sum(x) / i  

plt.plot(n, x_mean)

    
```

Now we check with Cauchy

```python
max_n = 10000
x_mean = np.zeros(max_n)
n = np.linspace(1, max_n, max_n)

for i in range(max_n):
    x = cauchy.rvs(size = i)
    x_mean[i] = np.sum(x) / i  

plt.plot(n, x_mean)
```

What i see in the cauchy distribution is that it is very erratic, the mean does not converge. This is because the Cauchy distribution has an undefined mean (the integral for the expected value does not converge). Also, law of large numbers does not apply to distributions with infinite / undefined mean

### Exercise 10
Simulating the stock market. $Y_1, Y_2 ...$  are independent random variables with:

p($Y_i$ = -1) = 1/2 and p($Y_i$  = 1) = 1/2

$X_n = \sum_{i=1}^n Y_n$

We find that 
$E(X_n) = 0$ and $V(X_n) = n$

Now simulate $X_n$ and plot $X_n$ versus n for n = 1,2, ... ,10,000. Repeat
the whole simulation several times.

```python
max_n = 1000
x_n = np.zeros(max_n)
n = np.linspace(1, max_n, max_n)

for i in range(max_n):
    b = bernoulli.rvs(p=.5, size = i)
    x = np.where(b == 0, -1, 1)
    x_n[i] = np.sum(x)

plt.plot(n, x_n)
```

The variance increases as the simulation increases (variance is exactly n) , making it difficult to predict in the later future
